# Make predictions on 1.0s-DataSet

```python
# Create a function to import an image and resize it to be able to be used with our model
def load_and_prep_image(filename, img_shape=224, scale=True):
  """
  Reads in an image from filename, turns it into a tensor and reshapes into
  (224, 224, 3).

  Parameters
  ----------
  filename (str): string filename of target image
  img_shape (int): size to resize target image to, default 224
  scale (bool): whether to scale pixel values to range(0, 1), default True
  """
  # Read in the image
  img = tf.io.read_file(filename)
  # Decode it into a tensor
  img = tf.io.decode_image(img, channels=3)
  # Resize the image
  img = tf.image.resize(img, [img_shape, img_shape])
  if scale:
    # Rescale the image (get all values between 0 and 1)
    return img/255.
  else:
    return img
```

```python
# Make a function to predict on images and plot them (works with multi-class)
def pred_and_plot(model, filename, class_names):
  """
  Imports an image located at filename, makes a prediction on it with
  a trained model and plots the image with the predicted class as the title.
  """
  # Import the target image and preprocess it
  img = load_and_prep_image(filename)

  # Make a prediction
  pred = model.predict(tf.expand_dims(img, axis=0))

  # Get the predicted class
  if len(pred[0]) > 1: # check for multi-class
    pred_class = class_names[pred.argmax()] # if more than one output, take the max
  else:
    pred_class = class_names[int(tf.round(pred)[0][0])] # if only one output, round

  # Plot the image and predicted class
  plt.imshow(img)
  plt.title(f"Prediction: {pred_class}")
  plt.axis(False);
```

```python
import os
DATA_DIR = '/content/acc'
CLASS_NAMES = sorted(os.listdir(DATA_DIR))
CLASS_NAMES
```

```python
import matplotlib.pyplot as plt
pred_and_plot(pretrained_model, '/content/acc/over_yield/189-wave1-fourier.png', class_names=CLASS_NAMES)
```

# Evaluation on complete dataset

```python
BATCH_SIZE = 32
IMG_SIZE = (224, 224)
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
test_data = test_datagen.flow_from_directory(directory=DATA_DIR,
                                             batch_size=BATCH_SIZE,
                                             target_size=IMG_SIZE,
                                             class_mode='binary',
                                             shuffle=False)
```

```python
pretrained_model.evaluate(test_data)
```
